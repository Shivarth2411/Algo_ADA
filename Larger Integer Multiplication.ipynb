```python
def largeint(x,y):
    
    xstr = str(x)
    ystr = str(y)
    
    if len(xstr)==1 or len(ystr)==1:
        return x*y
    
    n=max(len(xstr),len(ystr))
    
    nhalf = n//2
    xhigh = int(xstr[:-nhalf] or '0')
    xlow = int(xstr[-nhalf:])
    yhigh = int(ystr[:-nhalf] or '0')
    ylow = int(ystr[-nhalf:])
    
    a = largeint(xhigh,yhigh)
    b = largeint(xlow,ylow)
    c = largeint(xhigh + xlow,yhigh + ylow)- a - b
    
    result = a*10**(2*nhalf) + c * 10**nhalf + b
    
    return result
```

```python
x = int(input("Enter the Number A = "))
y = int(input("Enter the Number B = "))

result = largeint(x,y)

print("Threfore Multiplication is A x B = ",result)
```
